```python
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
clq_3 = pd.read_pickle('./cliques_rmsd_dist3.pkl')
clq_3['rmsd'] = clq_3[2].str[0]
clq_3['dist'] = clq_3[2].str[1]
clq_3['cand'] = np.where(clq_3.rmsd < 0.15,1,0)
clq_3.head()
```

```python
clq_3.groupby('cand').dist.describe(percentiles=np.linspace(0,1,11)).T
```
